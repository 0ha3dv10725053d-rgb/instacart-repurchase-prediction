```python
import pandas as pd
df = pd.read_csv('/user_department_reorerd-1769690016696.csv')
ad = pd.read_csv('/low_prior.csv')
df.head(10)

```

```python
df2 = pd.merge(df, ad, on=['user_id',"department"], how='left')
df2.head()
```

```python
df3 = df2.drop(columns=['user_id','user_id:1',"department","dept_reorder_rate"])
df3.head()
```

```python
x = df3.drop(columns=["reordered"])
y = df3["reordered"]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np
model = LogisticRegression()
y= y.fillna(0)
x = np.nan_to_num(x)
model.fit(x,y)
```

```python
print("正解率:{:.3f}".format(model.score(x, y)))
```

```python
y_pred_proba = model.predict_proba(x)[:, 1]
print(y_pred_proba)
```

```python
from sklearn.metrics import roc_auc_score

auc = roc_auc_score(y, y_pred_proba)
print(f"AUC: {auc:.3f}")
```

```python
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt

fpr, tpr, thresholds = roc_curve(y, y_pred_proba)

plt.plot(fpr, tpr, label=f"AUC = {auc:.2f}")
plt.plot([0, 1], [0, 1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()
```
